```python
import pandas as pd
warnings.filterwarnings("ignore")
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to 'Car Model' column
new_dataset['Car Model'] = le.fit_transform(new_dataset['Car Model'])
new_dataset['Car Make'] = le.fit_transform(new_dataset['Car Make'])

# Print the updated dataset
print(new_dataset)
```

```python
import numpy as np

# Assign 'EV' for 'Engine Size (L)' equal to 'EV', else assign 'Fuel'
new_dataset['Vehicle Type'] = np.where(new_dataset['Engine Size (L)'] == 'EV', 'EV', 'Fuel')

# Assign 'Hybrid' for 'Engine Size (L)' equal to 'Hybrid', else retain the existing value
new_dataset['Vehicle Type'] = np.where(new_dataset['Engine Size (L)'] == 'Hybrid', 'Hybrid', new_dataset['Vehicle Type'])

new_dataset['Engine Size (L)'] = new_dataset['Engine Size (L)'].replace('EV', 0)
new_dataset['Engine Size (L)'] = new_dataset['Engine Size (L)'].replace('Hybrid', 00)

# Check the result
new_dataset['Vehicle Type'].value_counts()

```
