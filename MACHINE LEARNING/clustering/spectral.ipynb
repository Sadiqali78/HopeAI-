```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("cluster.csv")
datasett=dataset.iloc[:,[3,4]]
datasett
```

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

# Assuming `datasett` is a pandas DataFrame
# Perform spectral clustering
sc = SpectralClustering(n_clusters=3, assign_labels='discretize', random_state=0)
y_SC = sc.fit_predict(datasett)

# Scatter plot for the clusters
plt.scatter(datasett.iloc[y_SC == 0, 0], datasett.iloc[y_SC == 0, 1], 
            s=50, c='red', label='Cluster 1')
plt.scatter(datasett.iloc[y_SC == 1, 0], datasett.iloc[y_SC == 1, 1], 
            s=50, c='blue', label='Cluster 2')
plt.scatter(datasett.iloc[y_SC == 2, 0], datasett.iloc[y_SC == 2, 1], 
            s=50, c='yellow', label='Cluster 3')

# Adding title and labels
plt.title('Spectral Clustering Results')
plt.xlabel('Feature 1')  # Replace with meaningful feature name if available
plt.ylabel('Feature 2')  # Replace with meaningful feature name if available
plt.legend()
plt.show()
```
